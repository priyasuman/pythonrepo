```python

##Given a list of numbers, count and print how many are even vs odd via a for loop

numbers = [10, 15, 22, 33, 40, 55, 60]

even_count = 0
odd_count = 0

for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Even numbers count:", even_count)
print("Odd numbers count:", odd_count)
```

```python
## Fibonacci series in python
n = int(input("Enter the number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")

for i in range(n):
    print(a)

    c = a + b
    a = b
    b = c
```

##4.Loop Control Statements

```python
## Break statement terminates the loop
for i in range(10):
    if i==5:
        break
    print(i)
    
   
```

```python
##Continue: Skips the iteration and continues the loop
for i in range(10):
    if i==5:
        continue
    print(i)
```

```python
## pass: placeholder statement. It just checks if the syntax is correct and does nothing

for i in range(10):
    if i==5:
        pass
    print(i)
```

```python
##WAPP program that prompts the user to enter two numbers and performs bitwise AND, OR, and XOR operations on them. Print out the results of each operation
# Ask the user to enter two integers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Perform bitwise operations
bitwise_and = num1 & num2
bitwise_or = num1 | num2
bitwise_xor = num1 ^ num2

# Print the results
print("Bitwise AND:", bitwise_and)
print("Bitwise OR:", bitwise_or)
print("Bitwise XOR:", bitwise_xor)
```

```python
##WAPP that checks whether a given year is a leap year or not. Prompt the user to enter a year and then determine if it is a leap year using the following criteria:  A year is a leap year if it is divisible by 4 but not divisible by 100, except when it is also divisible by 400.


year = int(input("Enter a year: "))

# Check if the year is a leap year
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.") ## % gives the remainder after division.
else:
    print(year, "is not a leap year.")
```

```python
##WAPP that prompts the user to enter a sequence of numbers. Use a 'while' loop to iterate through the sequence, and if the user enters a negative number, break out of the loop. Print the sum of all the positive numbers entered before the negative number

# Initialize sum
total = 0

# Loop until a negative number is entered
while True:
    num = int(input("Enter a number (negative number to stop): "))

    if num < 0:
        break

    total += num

# Display the sum
print("Sum of positive numbers entered:", total)
```
